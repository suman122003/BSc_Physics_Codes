# Legendre Polynomials

$$ (1-x^2)y'' - 2xy' + n(n+1)y = 0 $$

$$ P_n(x) = \sum_{k=0}^{m} (-1)^k \frac{(2n-2k)!}{2^n k! (n-k)!(n-2k)!} x^{n-2k} $$

For $n=even$, $m=\frac{n}{2}$ and for $n=odd$, $m=\frac{n-1}{2}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre as P

x = np.linspace(-1,1,100)
P0 = P(0)(x)
P1 = P(1)(x)
P2 = P(2)(x)
P3 = P(3)(x)
P4 = P(4)(x)
fig, axes = plt.subplots(2,3, figsize=(12,6))
ax1 = axes[0][0]
ax1.plot(x,P0, label='$P_0(x)$')
ax1.legend()
ax2 = axes[0][1]
ax2.plot(x,P1, 'orange', label='$P_1(x)$')
ax2.legend()
ax3 = axes[0][2]
ax3.plot(x,P2, 'green', label='$P_2(x)$')
ax3.legend()
ax4 = axes[1][0]
ax4.plot(x,P3, 'red', label='$P_3(x)$')
ax4.legend()
ax5 = axes[1][1]
ax5.plot(x,P4, 'violet', label='$P_4(x)$')
ax5.legend()
ax6 = axes[1][2]
ax6.plot(x,P0, label='$P_0(x)$')
ax6.plot(x,P1, label='$P_1(x)$')
ax6.plot(x,P2, label='$P_2(x)$')
ax6.plot(x,P3, label='$P_3(x)$')
ax6.plot(x,P4, label='$P_4(x)$')
ax6.legend()
plt.show()
```

## Orthogonality of Legendre Polynomials

$$ \int_{-1}^{+1} P_m(x)P_n(x) dx = \frac{2}{2n+1} \delta_{mn} $$

```python
import numpy as np
from scipy.special import legendre

def pmpn(m,n):
    pm = legendre(m)
    pn = legendre(n)
    return pm*pn

# Integration by Simpson's 1/3rd Rule
def simp13x(f,a,b,n):
    h = float(b-a)/n
    x0 = np.arange(a+h,b,2*h)
    xe = np.arange(a+2*h,b,2*h)
    val = h/3*(f2(a) + 4*sum(f2(x0)) + 2*sum(f2(xe)) + f2(b))
    return val

# Check Orthogonality for different values of m and n.
m = 3
n = 3

f2 = pmpn(m,n)
intg = simp13x(f2,-1,1,1000)  # integration

if m==n:
    dmn = 1
else:
    dmn = 0
res = (2/(2*n + 1))* dmn  # result
print('m =' , m, 'n =', n, '\n', 'RHS =', intg, '\t', 'LHS =', res)  # compare the values of intg and res
```
