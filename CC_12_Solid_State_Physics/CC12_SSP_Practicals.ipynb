## Hall Coefficient of a Semiconductor sample

Working formula:
$$ R_H = \frac{V_H \, t}{I_H \, H} $$

Carrier density:
$$ n = \frac{1}{R_H \, q} $$

### Data (18/12/2023)

```python
expt10_1 = pd.read_excel('CC_12_Practicals_data.xlsx', sheet_name='10.1')
expt10_2 = pd.read_excel('CC_12_Practicals_data.xlsx', sheet_name='10.2')
expt10_3 = pd.read_excel('CC_12_Practicals_data.xlsx', sheet_name='10.3')
display('table 1',expt10_1.head(),'table 2',expt10_2.head(),'table 3',expt10_3.head())
```

Table 2

```python
t = 0.67e-1  # given

table2H = expt10_2['H'].to_numpy()
table2VH = expt10_2['VH'].to_numpy()
table2IH = expt10_2['IH'].to_numpy()
# table2RH = table2VH*t/ (table2IH*table2H) # Hall coefficients
# table2n = 1/(table2RH*q)
# display(table2RH, np.mean(table2RH), table2n, np.mean(table2n))

f1 = lambda x, a,b: a*x +b
popt, pcov = curve_fit(f1, table2H, table2VH, p0=[2.5,0.5])
aopt, bopt = popt[0], popt[1]
table2VHopt = f1(table2H, aopt,bopt)
RHopt1 = aopt*t/table2IH[0]
print(f'Optimal parameters: {aopt, bopt}. Value of Hall coefficient from this table is {RHopt1}.')
```

```python
# px.scatter(expt10_2, x='H',y='VH', hover_data='IH')
```

```python
plt.figure(figsize=(9,5))
plt.title('Magnetic field (H) vs Hall Voltage (VH) at constant Hall current (IH)\n')
sns.scatterplot(data=expt10_2, x='H', y='VH', hue='IH', palette='Set2')
# sns.lineplot(data=expt10_2, x='H', y='VH', hue='IH', palette='Accent_r', linewidth=0.5)
plt.plot(table2H, table2VHopt, lw=0.8)
plt.xticks(np.arange(0, 16, 1))
plt.yticks(np.arange(0, 35, 5))
sns.despine()
plt.savefig('hall effect table 2', dpi=600)
plt.show()
```

Table 3

```python
t = 0.67e-1  # given
table3H = expt10_3['H'].to_numpy()
table3VH = expt10_3['VH'].to_numpy()
table3IH = expt10_3['IH'].to_numpy()

f1 = lambda x, a,b: a*x +b
popt, pcov = curve_fit(f1, table3IH, table3VH, p0=[2.5,0.5])
aopt, bopt = popt[0], popt[1]
table3VHopt = f1(table3IH, aopt,bopt)
RHopt2 = aopt*t/table3H[0]
print(f'Optimal parameters: {aopt, bopt}. Value of Hall coefficient from this table is {RHopt2}.')
```

```python
plt.figure(figsize=(8,6))
plt.title('Hall current (IH) vs Hall Voltage (VH) at constant Magnetic field (H)')
sns.scatterplot(data=expt10_3, x='IH', y='VH', hue='H', palette='Set2')
# sns.lineplot(data=expt10_3, x='IH', y='VH', hue='H', palette='Accent_r', linewidth=0.5)
plt.plot(table3IH, table3VHopt, lw=0.8)
plt.xticks(np.arange(0, 6, 0.5))
plt.yticks(np.arange(0, 45, 5))
sns.despine()
plt.savefig('hall effect table 3', dpi=600)
plt.show()
```

Calculations

```python
RHmean = np.mean([RHopt1, RHopt2])
print(f'Value of RH: from table 2: {RHopt1}, from table 3: {RHopt2}, mean: {RHmean}.')
q = 1.6e-19
n = 1/(RHmean*q)
print(f'Hall coefficient: {RHmean}. Carrier density: {n}.')
```
