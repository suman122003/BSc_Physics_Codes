# Barrier Potential

## 1D

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import hbar

# Constants
L = 10.0  # Length of the 1D domain
N = 500  # Number of grid points
x = np.linspace(0, L, N)  # Define the x-axis

# Define the finite barrier potential
barrier_height = 20.0  # Higher barrier
barrier_width = 2.0
barrier_position = L / 3  # Position of the barrier

# Potential function for a 1D finite barrier potential
def barrier_potential(x):
    return np.where(
        (x >= barrier_position) & (x <= barrier_position + barrier_width),
        barrier_height,
        0.0
    )

# Constants for the Schrödinger equation
mass = 1.0  # Particle mass
dx = x[1] - x[0]
T = (-hbar**2 / (2 * mass)) * (np.diag(-2.0 / dx**2 * np.ones(N)) + np.diag(1.0 / dx**2 * np.ones(N - 1), 1) + np.diag(1.0 / dx**2 * np.ones(N - 1), -1))
V = barrier_potential(x)
H = T + np.diag(V * np.ones(N))

# Solve the Schrödinger equation
eigvals, eigvecs = np.linalg.eigh(H)

# Plot the wavefunction for the lowest energy state
wavefunction = eigvecs[:, 0]
plt.plot(x, wavefunction, label='Wavefunction (E < V)')
plt.axvline(x=barrier_position, color='red', linestyle='--', label='Left End of Barrier')
plt.axvline(x=barrier_position + barrier_width, color='green', linestyle='--', label='Right End of Barrier')
plt.xlabel('Position (x)')
plt.ylabel('Wavefunction (ψ)')
plt.title('Wavefunction for a Particle with Energy E < V in a Finite 1D Barrier Potential')
plt.legend()
plt.grid()
plt.show()
```
